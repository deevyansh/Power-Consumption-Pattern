```python
# Importing the data
import pandas as pd
covariance_df=pd.read_csv('NO_covariance.csv')
```

```python
# Days of the month on which there is high covariance
high_correlation_days = pd.DataFrame(index=range(36), columns=['date', 'month', 'is_correlated_with_DK1'])
for i in range(0, 365):
    date_index = i * 24
    count = 0
    date = covariance_df['date'].iloc[date_index]
    month = covariance_df['month'].iloc[date_index]
    for j in range(0, 24):
      hour_index = date_index + j
      if covariance_df['covariance_value_with_DK1'].iloc[hour_index] > 0.5:
          count += 1
    if count > 12:
          is_correlated_DK1 = 1
    else:
          is_correlated_DK1 = 0
    count=0
    for j in range(0, 24):
          hour_index = date_index + j
          if covariance_df['covariance_value_with_DK2'].iloc[hour_index] > 0.5:
              count += 1
    if count > 12:
          is_correlated_DK2 = 1
    else:
          is_correlated_DK2 = 0
    is_correlated=0
    if (is_correlated_DK1==1 and is_correlated_DK2==1):
        is_correlated=1
    high_correlation_days.loc[i, 'date'] = date
    high_correlation_days.loc[i, 'month'] = month
    high_correlation_days.loc[i, 'is_correlated_with_DK1'] = is_correlated_DK1
    high_correlation_days.loc[i, 'is_correlated_with_DK2'] = is_correlated_DK2
    high_correlation_days.loc[i, 'with both'] = is_correlated
    if (is_correlated==1):
      print(high_correlation_days['month'].iloc[i],high_correlation_days['date'].iloc[i])
```

```python
import pandas as pd

high_correlation_hours = pd.DataFrame(columns=['hours', 'month', 'is_correlated_with_DK1','is_correlated_with_DK2'])

for i in range(1, 13):
    for j in range(0, 24):
        new_df_hour = covariance_df[(covariance_df['month'] == i) & (covariance_df['hour'] == j)]
        count = 0
        final_index=((24*(i-1))+j)
        high_correlation_hours.loc[final_index, 'month'] = i
        high_correlation_hours.loc[final_index, 'hours'] = j
        for index, row in new_df_hour.iterrows():
            if row['covariance_value_with_DK1'] > 0.5:
                count = count + 1
        if count > 7:
          high_correlation_hours.loc[final_index, 'is_correlated_with_DK1'] = 1
        else:
          high_correlation_hours.loc[final_index, 'is_correlated_with_DK1'] = 0
        count=0
        for index, row in new_df_hour.iterrows():
            if row['covariance_value_with_DK2'] > 0.5:
                count = count + 1
        if count > 7:
          high_correlation_hours.loc[final_index, 'is_correlated_with_DK2'] = 1
        else:
          high_correlation_hours.loc[final_index, 'is_correlated_with_DK2'] = 0
        if (high_correlation_hours.iloc[final_index]['is_correlated_with_DK1']==1 & high_correlation_hours.iloc[final_index]['is_correlated_with_DK2']==1):
          print(i," ", j)
high_correlation_hours
```

```python
#1.For 365 days(most recent year), For DK1,DK2 get daily averages(total/24) - 12 graphs (for every month)
#2.x = 25 percent of daily average for that day. for that 24 hours check if it greater than x.
#3.Table made is table of peak hours.
#4.Check if correlation > +0.5(+0.25) for those hours.
DK1_mixed=pd.read_csv('DK1_mixed.csv')

DAILY_AVERAGE=pd.DataFrame(columns=['2023_values','Date','Month'])
for i in range (0,365):
  sum=0
  for j in range (0,24):
    sum=sum+DK1_mixed['2023_values'].iloc[i*24+j]
  DAILY_AVERAGE.loc[i]=({'2023_values':sum/24,'Date': DK1_mixed['date'].iloc[i*24],'Month': DK1_mixed['month'].iloc[i*24]})
print(DAILY_AVERAGE)
```

```python
DAILY_AVERAGE['TotalLoad+25%']=DK1_mixed['2023_values']+(DK1_mixed['2023_values']/4)
DAILY_AVERAGE
```

```python
PEAK_HOURS=pd.DataFrame(columns=['hour','date','month','covariance','value'])
num=0
for i in range (0,365):
  for j in range (0,24):
    if(DK1_mixed['2023_values'].iloc[24*i+j]>DAILY_AVERAGE['TotalLoad+25%'].iloc[i]):
      PEAK_HOURS.loc[num]=({'hour':DK1_mixed['hour'].iloc[24*i+j],'date':DK1_mixed['date'].iloc[24*i+j],'month':DK1_mixed['month'].iloc[24*i+j],'covariance':0})
      num=num+1
print(PEAK_HOURS)
```

```python
import pandas as pd
for i in range(0, 901):
    mask = (covariance_df['month'] == PEAK_HOURS['month'].iloc[i]) & \
           (covariance_df['hour'] == PEAK_HOURS['hour'].iloc[i]) & \
           (covariance_df['date'] == PEAK_HOURS['date'].iloc[i])
    PEAK_HOURS.loc[i, 'covariance'] = covariance_df[mask]['covariance_value_with_DK1'].iloc[0]
PEAK_HOURS
```

```python
PEAK_HOURS_GOOD_COVARIANCE=pd.DataFrame()
for i in range (0,901):
  if(PEAK_HOURS['covariance'].iloc[i]>0.5):
    PEAK_HOURS_GOOD_COVARIANCE=pd.concat([PEAK_HOURS_GOOD_COVARIANCE,PEAK_HOURS.iloc[i]],axis=1)
PEAK_HOURS_GOOD_COVARIANCE=PEAK_HOURS_GOOD_COVARIANCE.transpose()
PEAK_HOURS_GOOD_COVARIANCE=PEAK_HOURS_GOOD_COVARIANCE.reset_index()
PEAK_HOURS_GOOD_COVARIANCE
```

```python
df_mixed=pd.read_csv('NO_mixed.csv')
for i in range (0,35):
  hour=PEAK_HOURS_GOOD_COVARIANCE.loc[i,'hour']
  date=PEAK_HOURS_GOOD_COVARIANCE.loc[i,'date']
  month=PEAK_HOURS_GOOD_COVARIANCE.loc[i,'month']
  value=df_mixed[(df_mixed['hour']==hour) & (df_mixed['month']==month) & (df_mixed['date']==date)]['values_2023']
  PEAK_HOURS_GOOD_COVARIANCE.loc[i,'value']=0.1*value.iloc[0]
PEAK_HOURS_GOOD_COVARIANCE.to_csv('Peak hour NO_DK1.csv')
```

```python
# Importing the file
DK2_mixed=pd.read_csv('DK2_mixed.csv')

DAILY_AVERAGE=pd.DataFrame(columns=['TotalLoad','Date','Month'])
for i in range (0,365):
  sum=0
  for j in range (0,24):
    sum=sum+DK2_mixed['2023_values'].iloc[i*24+j]
  DAILY_AVERAGE.loc[i]=({'TotalLoad':sum/24,'Date': DK2_mixed['date'].iloc[i*24],'Month': DK2_mixed['month'].iloc[i*24]})
print(DAILY_AVERAGE)
```

```python
DAILY_AVERAGE['TotalLoad+25%']=DK2_mixed['2023_values']+(DK2_mixed['2023_values']/4)
DAILY_AVERAGE
```

```python
#Peak Hours Finding
PEAK_HOURS=pd.DataFrame(columns=['hour','date','month','covariance','value'])
num=0
for i in range (0,365):
  for j in range (0,24):
    if(DK2_mixed['2023_values'].iloc[24*i+j]>DAILY_AVERAGE['TotalLoad+25%'].iloc[i]):
      PEAK_HOURS.loc[num]=({'hour':DK2_mixed['hour'].iloc[24*i+j],'date':DK2_mixed['date'].iloc[24*i+j],'month':DK2_mixed['month'].iloc[24*i+j],'covariance':0})
      num=num+1
print(PEAK_HOURS)
```

```python
#Adding the covariance value of peak hours
import pandas as pd

# Assuming PEAK_HOURS and covariance_df DataFrames exist

# Convert columns to integer type
PEAK_HOURS['month'] = PEAK_HOURS['month'].astype(int)
PEAK_HOURS['hour'] = PEAK_HOURS['hour'].astype(int)
PEAK_HOURS['date'] = PEAK_HOURS['date'].astype(int)

# Correct the assignment using loc
for i in range(0, 720):
    mask = (covariance_df['month'] == PEAK_HOURS['month'].iloc[i]) & \
           (covariance_df['hour'] == PEAK_HOURS['hour'].iloc[i]) & \
           (covariance_df['date'] == PEAK_HOURS['date'].iloc[i])
    # Use loc to set the value
    PEAK_HOURS.loc[i, 'covariance'] = covariance_df[mask]['covariance_value_with_DK2'].iloc[0]
# Display the DataFrame
PEAK_HOURS
```

```python
#Peak Hours with good covariance value
PEAK_HOURS_GOOD_COVARIANCE=pd.DataFrame()
for i in range (0,720):
  if(PEAK_HOURS['covariance'].iloc[i]>0.7):
    PEAK_HOURS_GOOD_COVARIANCE=pd.concat([PEAK_HOURS_GOOD_COVARIANCE,PEAK_HOURS.iloc[i]],axis=1)
PEAK_HOURS_GOOD_COVARIANCE = PEAK_HOURS_GOOD_COVARIANCE.transpose()
PEAK_HOURS_GOOD_COVARIANCE= PEAK_HOURS_GOOD_COVARIANCE.reset_index(drop=True)
PEAK_HOURS_GOOD_COVARIANCE
PEAK_HOURS_GOOD_COVARIANCE.to_csv('Peak hour with good covariance in DK2.csv')
```
